## Task 6

The other numeric column is `event_year`.
- Determine the earliest and latest years in the DataFrame

```python
earliest_year = laptops['event_year'].min()
latest_year = laptops['event_year'].max()

print(f'Earliest year: {earliest_year}')
print(f'Latest year: {latest_year}')
```

## Task 7

Let's also take a look at `event_year` as a categorical column.
- Preview the number of laptops in the DataFrame for each `event_year`

```python
event_year_counts = laptops['event_year'].value_counts()
event_year_counts
```

## Task 8

Let's focus on the problems causing people to bring in laptops for repair.
- Preview the number of problems in the DataFrame

```python
problem_counts = laptops['problem'].value_counts()
problem_counts
```

## Task 9

Power and battery issues are pretty common, but what percentage of the data do they represent?

- Modify the **Task 8** and preview out the percentages instead of counts

```python
problem_percentages = laptops['problem'].value_counts(normalize=True) * 100
problem_percentages
```
