# Electric Vehicle Market Size Analysis

This report offers insights and visualizations and conducts a market size analysis of the electric vehicle (EV) industry. But what exactly is market size analysis? It helps businesses gauge the potential of a market, evaluate competition, and identify growth opportunities.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Global Settings

- Set the Seaborn style to `whitegrid` for better visualization.
- Dataset URL

```python
# Seaborn style
sns.set_style("whitegrid")

# Dataset URL
ds_url = "https://raw.githubusercontent.com/cogxen/datasets/main/market-size-of-evs/dataset.csv"
```

## Data Preprocessing
- Load the dataset
- Check for missing values
- Drop missing values

### Load the Dataset

- Display the first 20 rows of the dataset to understand the data and to gain insights on the columns and values.

```python
ds = pd.read_csv(ds_url)
ds.head(20)
```

As you can see, the data is based on the electric vehicle population in the United States. 

```python
ds.info()
```

### Check for Missing Values

- Check for missing values in the dataset to ensure data quality and completeness.

```python
ds.isnull().sum()
```

### Drop Missing Values

- Remove those missing data or values to ensure the quality of the data.

```python
# Drop the missing values
ds = ds.dropna()

# Check if there are any missing values left, just making sure.
ds.isnull().sum()
```

## Objectives

For the objectives of this analysis, we will explore the following areas:
1. **Electric Vehicle Adoption Over time** - Analyze the yearly growth in the population of electric vehicles.
2. **Geographic Distribution** - Understand how electric vehicles are distributed across various regions, including by county and city.
3. **Electric Vehicle Type** - Examine the distribution of electric vehicles by their type.
4. **Make and Model Popularity** - Identify the most popular makes and models of electric vehicles among registered users.

### Electric Vehicle Adoption Over Time
Analyze the adoption or growth of the electric vehicle population over time by visualizing the number of electric vehicles registered by model year. This will give us an insight on how the electric vehicle market has evolved over the years.

```python
# Plot the number of electric vehicles by model year
plt.figure(figsize=(12, 6))
ds_adoption_by_year = ds["Model Year"].value_counts().sort_index()
sns.barplot(x=ds_adoption_by_year.index, y=ds_adoption_by_year.values, palette="viridis")
plt.title("Electric Vehicle Adoption Over Time")
plt.xlabel("Model Year")
plt.ylabel("Number of Registered Electric Vehicles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Based on the visualization, we can see that the number of registered electric vehicles has been increasing over the years. The adoption of electric vehicles has been growing, especially in recent years. The year 2023 shows a drastic increase in the number of registered electric vehicles, indicating a significant growth in the electric vehicle market.

### Geographic Distribution

Understand how electric vehicles are distributed across various regions, including by county and city. This will provide insights into the geographic distribution of electric vehicles and identify areas with high adoption rates.

#### By County Level

- Analyze the distribution of electric vehicles by the top 10 counties with the highest number of registered electric vehicles.

```python
# Distribution by the top 10 counties
ds_distribution_by_top_10_county = ds["County"].value_counts().head(10)
top_counties = ds_distribution_by_top_10_county.index

# Plot the distribution of electric vehicles by county
plt.figure(figsize=(12, 6))
sns.barplot(x=ds_distribution_by_top_10_county.values, y=top_counties, palette="viridis")
plt.title("Top 10 Counties with Highest Number of Registered Electric Vehicles")
plt.xlabel("Number of Registered Electric Vehicles")
plt.ylabel("County")
plt.tight_layout()
plt.show()
```

#### By City Level

- Analyze the distribution of electric vehicles by the top 10 cities with the highest number of registered electric vehicles.

```python
# Distribution by the top 10 cities
ds_distribution_by_city = ds["City"].value_counts().head(10)
top_cities = ds_distribution_by_city.index

# Plot the distribution of electric vehicles by city
plt.figure(figsize=(12, 6))
sns.barplot(x=ds_distribution_by_city.values, y=top_cities, palette="viridis")
plt.title("Top 10 Cities with Highest Number of Registered Electric Vehicles")
plt.xlabel("Number of Registered Electric Vehicles")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

These visualizations show the top 10 counties and cities with the highest number of registered electric vehicles. The data provides insights into the geographic distribution of electric vehicles, highlighting areas with high adoption rates.

### Electric Vehicle Type

Examine the types of electric vehicles in the dataset. Let's understand the breakdown between Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs), provide insights to the consumer preference and the adoption patterns of different electric vehicle types.

```python
# Getting the count of different types of electric vehicles
ds_vehicle_type = ds["Electric Vehicle Type"].value_counts()

# Plot the distribution of electric vehicle types
plt.figure(figsize=(8, 6))
sns.barplot(x=ds_vehicle_type.index, y=ds_vehicle_type.values, palette="viridis")
plt.title("Distribution of Electric Vehicle Types")
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Number of Registered Electric Vehicles")
plt.tight_layout()
plt.show()
```

It visualized that the majority of registered electric vehicles are Battery Electric Vehicles (BEVs), indicating a higher adoption rate of BEVs compared to Plug-in Hybrid Electric Vehicles (PHEVs).

### Make and Model Popularity

Identify the most popular makes and models of electric vehicles among registered users. This would help us to identify which manufacturers and models are leading the electric vehicle market and would dominate the electric vehicle market.

```python
# Getting the count of electric vehicles by make
ds_make_distribution = ds["Make"].value_counts()

# Plot the distribution of electric vehicle makes
plt.figure(figsize=(12, 6))
sns.barplot(x=ds_make_distribution.head(10).values, y=ds_make_distribution.head(10).index, palette="viridis")
plt.title("Top 10 Most Popular Electric Vehicle Makes")
plt.xlabel("Number of Registered Electric Vehicles")
plt.ylabel("Make")
plt.tight_layout()
plt.show()
```

Based on the visualizations, we can see the most popular electric vehicle makes and models among the registered population. 

### Electric Range Analysis

- Electric range is the distance an electric vehicle can travel on a single charge

Investigate the electric range of vehicles to track the technological advancements over time. We will visualize the distribution of electric range among electric vehicles to understand how the electric vehicle technology has evolved over time. We need to explore the electric range of electric vehicles, which is a crucial factor for consumers when choosing an electric vehicle and for analyzing the market size of electric vehicles. 

```python
# Plot the distribution of electric range
plt.figure(figsize=(12, 6))
sns.histplot(ds["Electric Range"], kde=True, color="skyblue", bins=30)
plt.title("Distribution of Electric Range of Electric Vehicles")
plt.xlabel("Electric Range (miles)")
plt.ylabel("Number of Electric Vehicles")
plt.axvline(ds["Electric Range"].mean(), color='red', linestyle='dashed', linewidth=1, label=f"Mean Range: {ds['Electric Range'].mean():.2f}")
plt.tight_layout()
plt.legend()
plt.show()
```

The visualization shows the distribution of electric range among electric vehicles. The mean electric range of electric vehicles is indicated by the red dashed line. The data provides insights into the electric range of electric vehicles, highlighting the advancements in battery technology and the increasing range of electric vehicles. The key observations from the analysis are:

- This is a high frequency of electric vehicles with a low electric range with a significant peak around just before 50 miles.
- The distribution is right-skewed, with a long tail towards higher electric ranges. Although the most number of vehicles with a low electric range, there are electric vehicles with a higher electric range.
- The mean electric range of electric vehicles is approximately 58.83 miles, indicating that the average electric vehicle can travel around 58.83 miles on a single charge.
- Despite the advancements in battery technology, there are still electric vehicles with a low electric range, which may impact consumer adoption and market growth.

### Estimate Market Size

Finally, let's estimate the market size of electric vehicles based on the dataset. The market size of electric vehicles is an important metric for businesses to understand the potential of the electric vehicle market and identify growth opportunities also this insights provide how advancements in battery technology have improved the electric range of electric vehicles.

```python
# Estimate the market size of electric vehicles by model year
ds_average_range_by_year = ds.groupby("Model Year")["Electric Range"].mean().reset_index()

# Plot the estimated market size of electric vehicles
plt.figure(figsize=(12, 6))
sns.lineplot(x=ds_average_range_by_year["Model Year"], y=ds_average_range_by_year["Electric Range"], marker="o", color="skyblue")
plt.title("Estimated Market Size of Electric Vehicles by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Average Electric Range (miles)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Based on the graph, it shows that the progression of the average electric range of vehicles from around the year 2000 to 2024. The data provides insights into how advancements in battery technology have improved the electric range of electric vehicles over time. The average electric range of electric vehicles has increased significantly over the years, indicating the growth of the electric vehicle market and the potential for further expansion. This analysis helps businesses understand the market size of electric vehicles and identify opportunities for growth and innovation. So here are the key takeaways from the analysis:
- There is a general upward trend in the average electric range of electric vehicles over time, indicating advancements in battery technology and improvements in electric vehicle performance.
- There's a noticeable peak around the year 2020 when the average electric range of electric vehicles increased significantly, showing a major shift in the electric vehicle market.
- Year 2020, there's a significant drop in the average electric range of electric vehicles, which may be due to the introduction of new models with varying electric ranges, affecting the overall average, the data that we have might be limited to certain models or incomplete, or it reflects the introduction of several lower-range models.
- Lastly, as you can see on the sharp descent in the year 2024, there's a quite slight recovery in the average range in the year 2024, which may indicate the introduction of new models with higher electric ranges or improvements in battery technology.

### Electric Vehicle Ranges Vary Among the Top Manufacturers and Models

- Analyze the electric range of the top 10 electric vehicle models to understand the variations in electric range among different manufacturers and models.

```python
# Calculate the average electric range by model
ds_average_range_by_model = ds.groupby("Model")["Electric Range"].mean().sort_values(ascending=False).head(10)

# Plot the average electric range by model
plt.figure(figsize=(12, 6))
sns.barplot(x=ds_average_range_by_model.values, y=ds_average_range_by_model.index, palette="viridis")
plt.title("Top 10 Electric Vehicle Models with Highest Average Electric Range")
plt.xlabel("Average Electric Range (miles)")
plt.ylabel("Model")
plt.tight_layout()
plt.show()
```

Based on the visualization, we can see the top 10 electric vehicle models with the highest average electric range. The data provides insights into the variations in electric range among different manufacturers and models, highlighting the advancements in battery technology and the increasing range of electric vehicles. The key observations from the analysis are:

- The top 10 electric vehicle models with the highest average electric range have a wide range of electric ranges, indicating the diversity in electric vehicle technology and performance.
- The average electric range of electric vehicles varies among different manufacturers and models, with some models offering significantly higher electric ranges than others.
- The data provides insights into the competitive landscape of the electric vehicle market and the advancements in electric vehicle technology.

### Market Size Analysis of EVs in the US

- Estimate the market size of electric vehicles in the United States based on the dataset. This would provide insights into the potential of the electric vehicle market in the US and help businesses understand the competitive landscape and identify growth opportunities.

```python
# Estimate the market size of electric vehicles in the US
market_size_us = ds["Electric Vehicle Type"].count()

# Plot the estimated market size of electric vehicles in the US
plt.figure(figsize=(6, 6))
plt.pie([market_size_us, 1000000-market_size_us], labels=["Electric Vehicles", "Non-Electric Vehicles"], autopct='%1.1f%%', colors=["skyblue", "lightcoral"])
plt.title("Market Size of Electric Vehicles in the US")
plt.tight_layout()
plt.show()
```
